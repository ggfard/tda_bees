```python
import matplotlib.pyplot as plt
```

```python
algs = ['Agglomerative\n(norm)', 'Kmeans\n(norm)',
        'Agglomerative\n(norm + time)', 'Kmeans\n(norm + time)\n', 
        'Changepoint\n(norm)\n', 
        'Agglomerative\n(pca + time)', 'Kmeans\n(pca + time)',
        'Changepoint\n(pca)'
       ]

x_ticks = [i*2 for i in range(len(algs))]

#hard coded from spreadsheet
means = [492.667, 454.167, 115, 58.333, 71.333, 44.667, 56.167, 32]
stdvs = [86.777, 185.046, 72.036, 87.157, 15.475, 140.018, 89.544, 60.120]
```

```python
#actually plotting
fig, ax = plt.subplots(figsize=(40, 15))

ax.scatter(x_ticks, means, color='black', s=200)
ax.errorbar(x_ticks, means, yerr=stdvs, ecolor='black', fmt='k.', elinewidth=2.5, capsize=20, capthick=2.5)
ax.set_xticks(x_ticks)
ax.set_xticklabels(algs, fontsize=40)
ax.set_ylim([-150, 650])
ax.set_yticklabels(['', '-100', '0', '100', '200', '300', '400', '500', '600'], fontsize=40)
ax.yaxis.grid(True) #turn on gray lines

ax.set_title(r'Average $t_{lag}$', fontsize=52)
ax.set_xlabel('Algorithm used', fontsize=50, labelpad=15)
ax.set_ylabel(r'$t_{lag}$ (seconds)', fontsize=50)

plt.savefig('../figures/delta_t_experimental.png', bbox_inches='tight', dpi=250)

plt.show()
```
