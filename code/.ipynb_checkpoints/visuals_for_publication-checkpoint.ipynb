```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_betti_norms(norms, kmeans_labels, agg_labels, cp_labels, fed_time, num_pts=1, savepath=None):
    '''Plots the Betti norms with annotations for the change points. 
    This function works for both experimental data. 
    
    Args:
        norms: The list of Betti norms
        kmeans_labels: The change points found by kmeans
        agg_labels: Like above, for agglomerative
        cp_labels: Like above, for Change Point
        fed_time: The time at which the fed bees are introduced
        num_pts: How many change points to plot, one or two
        savepath: If not none, the place t0 save the figure
    '''
    #truncate data to first 900 timesteps if only plotting one change point
    if num_pts == 1:
        norms = norms[:900]
        
    #some variables for style
    size = 40
    style = 'dashed'
    fed_color = 'black'
    norm_color = 'black'
    agg_c = '#FFB000'
    km_c = '#DC267F'
    cp_c = '#648FFF'
    
    #max height of vertical lines
    height = max(norms)
    
    #create plot
    plt.subplots(figsize=(32, 16))
    
    #plot norms
    plt.plot(norms, color=norm_color, lw=1)
    
    #plot fed bee annotation
    plt.vlines([fed_time], 0, height, colors=[fed_color], linestyles='solid', lw=3.5, label='Fed Bees')
    plt.annotate(s = str(fed_time), xy=(fed_time - 80, 350), color=fed_color, fontsize=size)
    
    #plot agglomerative annotation
    plt.vlines(agg_labels, 0, height, colors=[agg_c], linestyles='dashed', lw=3.5, label='Agglomerative')
    #could be one or two changepoints to label
    plt.annotate(s = str(agg_labels[0]), xy=(kmeans_labels[0] + 15, 410), color=agg_c, fontsize=size)
    if num_pts == 2:
        plt.annotate(s = str(agg_labels[1]), xy=(agg_labels[1] + 5, 500), color=agg_c, fontsize=size)
    
    #plot kmeans annotation
    plt.vlines(kmeans_labels, 0, height, colors=[km_c], linestyles=style, lw=3.5, label='K-Means')
    plt.annotate(s = str(kmeans_labels[0]), xy=(kmeans_labels[0] + 15, 450), color=km_c, fontsize=size)
    if num_pts == 2:
        plt.annotate(s = str(kmeans_labels[1]), xy=(kmeans_labels[1] + 5, 500), color=km_c, fontsize=size)
        
    #plot change point annotation
    plt.vlines(cp_labels, 0, height, colors=[cp_c], linestyles='dashed', lw=3.5, label='Change Point')
    plt.annotate(s = str(cp_labels[0]), xy=(kmeans_labels[0] + 15, 490), color=cp_c, fontsize=size)
    if num_pts == 2:
        plt.annotate(s = str(cp_labels[1]), xy=(cp_labels[1] + 5, 540), color=cp_c, fontsize=size)
    
    #figure out x and y ticks
    plt.xlim([0, len(norms)])
    plt.ylim([min(norms) - 20, max(norms) + 10])
    plt.xticks([i for i in range(0, len(norms), 200)] + [len(norms)], fontsize=size + 5)
    plt.yticks(fontsize=size + 5)
    
    #title, axis labels, legend
    plt.legend(loc='lower right', prop={'size': size})
    plt.ylabel(r'||$\vec{b}$||          ', fontsize=size + 10, rotation=0)
    plt.xlabel('Time', fontsize= size + 10)
    
    if savepath:
        plt.savefig(savepath, bbox_inches='tight')
    
    plt.show()
```

```python
#example plotting a single change point
num_pts = 1
norms_path = '../results/C0128_norms.txt'

#hard coded from the results saved to text files
fed_time = 430
kmeans_labels = [465]
agg_labels = [447]
cp_labels = [475]

#read norms
norms = np.loadtxt(norms_path, delimiter=',')[:,0] #second column is time annotation, which we discard here

#plot
plot_betti_norms(norms, kmeans_labels, agg_labels, cp_labels, fed_time, num_pts)
```

```python
#example for plotting two change points
num_pts = 2
number = 'C0128'
norms_path = '../crockers/' + number + '_betti_norms_annotated.txt'

#hard coded from the results saved to txt files
fed_time = 430
kmeans_labels = [510, 1066]
agg_labels = [447, 832]
cp_labels = [460, 1281]

#read norms
norms = np.loadtxt(norms_path, delimiter=',')[:,0] #second column is time annotation

#plot
plot_betti_norms(norms, kmeans_labels, agg_labels, cp_labels, fed_time, num_pts)
```

```python
def plot_model_betti_norms(norms, kmeans_labels, agg_labels, cp_labels, savepath=None):
    
    #style parameters
    fed_time = 400
    size = 40
    height = max(norms) + 3
    fed_color = 'black'
    agg_color = '#FFB000'
    km_c = '#DC267F'
    cp_c = '#648FFF'
    
    #we only ever use the first 900 timesteps of the model run
    norms = norms[:900]
    
    plt.subplots(figsize=(32, 16))
    
    plt.plot(norms, color='black', lw=1)
    
    #plot fed time annotation
    plt.vlines([fed_time], 0, height, colors=[fed_color], linestyles='solid', lw=3.5, label='Fed Bees')
    plt.annotate(text = str(fed_time), xy=(fed_time - 45, 60), color=fed_color, fontsize=size)
    
    #plot agglomerative annotation
    plt.vlines(agg_labels, 0, height, colors=[agg_color], linestyles='dashed', lw=3.5, label='Agglomerative')
    plt.annotate(text = str(agg_labels[0]), xy=(agg_labels[0] + 5, 80), color=agg_color, fontsize=size)

    #kmeans annotation
    plt.vlines(kmeans_labels, 0, height, colors=[km_c], linestyles= 'dashed', lw=3.5, label='K-Means')
    plt.annotate(text = str(kmeans_labels[0]), xy=(agg_labels[0] + 5, 60), color=km_c, fontsize=size)
        
    #change point annotation
    plt.vlines(cp_labels, 0, height - 8, colors=[cp_c], linestyles='dashed', lw=3.5, label='Change Point')
    plt.annotate(text = str(cp_labels[0]), xy=(agg_labels[0] + 5, 55), color=cp_c, fontsize=size)
    
    #figure out x and y ticks
    plt.xlim([0, len(norms)])
    plt.ylim([min(norms) - 10, max(norms) + 5])
    plt.xticks([i for i in range(0, len(norms), 200)] + [len(norms)], fontsize=size + 5)
    plt.yticks(fontsize=size + 5)
    
    #legend and axis labels
    plt.legend(loc='lower right', prop={'size': size})
    plt.ylabel(r'||$\vec{b}$||          ', fontsize=size + 10, rotation=0)
    plt.xlabel('Time', fontsize=size + 10)
    
    if savepath:
        plt.savefig(savepath, bbox_inches='tight')
    
    plt.show()
```

```python
#example plotting a model change point
number = 'positions_18'
norms_path = '../model_crockers/norms_' + number + '.txt'

kmeans_labels = [415]
agg_labels = [427]
cp_labels = [415]

norms = np.loadtxt(norms_path, delimiter=',')[:,0] #second column is time annotation which we don't need here

#plot
plot_model_betti_norms(norms, kmeans_labels, agg_labels, cp_labels)
```

```python
def plot_model_crocker(crocker, savepath=None):
    '''For creating CROCKER plots from model data.
    
    Args:
       crocker: The crocker plot read from a txt file
       savepath: If not none, the place to save the figure 
    '''
    levels=20
    
    #we always truncate the model runs at t=900
    matrix = np.array(crocker[:900,:])
    matrix = np.array(matrix).astype(float)
    matrix = matrix[::50,:]
    x,y  = np.meshgrid(50*np.arange(0,matrix.shape[0]), np.arange(0,20,1))

    plt.figure(dpi=300, figsize=(32,20))
    plt.contour(x, y, matrix[:,:].T, levels=levels, linewidths=.5, colors='k')
    hm = plt.contourf(x, y, matrix[:].T, levels=levels, cmap = plt.cm.jet_r)
    cbar = plt.colorbar(hm)
    cbar.ax.tick_params(labelsize=40)

    ax = plt.gca()
    ax.set_xlabel('Time (sec)', fontsize=50)
    ax.set_ylabel(r'$\epsilon$        ', fontsize=50, rotation=0)

    #annotate when fed bees are introduced. This is always t=400 for model data.
    plt.vlines([400], 0, levels - 1, colors=['white'], linestyles='solid', lw=2.5, label='Fed Bees')
    plt.legend(loc='best', prop={'size': 45})
    
    plt.xticks(fontsize=45)
    plt.yticks(fontsize=45)

    if savepath:
        plt.savefig(savepath, dpi=300, bbox_inches='tight')
    plt.show()
```

```python
#example for plotting a model CROCKER
number = 'positions_18'
crocker_path = '../model_crockers/crocker_' + number + '.txt'
crocker = np.loadtxt(crocker_path, delimiter=',')

plot_model_crocker(crocker)
```

```python
def plot_experiment_crocker(crocker, fed_time, truncate=True, savepath=None):
    '''Plots the CROCKER plots from experimental data. 
    
    Args:
        crocker: the CROCKER matrix
        fed_time: The time at which the fed bees are introduced
        truncate: If true, plot only the first 900 timesteps
        savepath: If not none, the place to save the figure 
    '''
    #reshape crocker for plotting
    if truncate:
        crocker = crocker[:900, :600]
    else:
        #we only plot first 600 filtration values for visual cleanliness
        crocker = crocker[:, :600]
    
    levels=20
    matrix = np.array(crocker)[:,:]
    matrix = np.array(matrix).astype(float)
    matrix = matrix[::,:]
    x,y  = np.meshgrid(np.arange(0,matrix.shape[0],1), np.arange(0,600,1))

    plt.figure(dpi=300, figsize=(20,12))

    hm = plt.contourf(x, y, matrix[:,:].T, levels=levels, cmap = plt.cm.jet_r)
    ax = plt.gca()
    ax.set_xlabel('Time(sec)', fontsize=40)
    ax.set_ylabel(r'$\epsilon$        ', fontsize=40, rotation=0)
    
    plt.vlines([fed_time], 0, 599, colors=['white'], linestyles='solid', label='Fed Bees')
    plt.legend(loc='best', prop={'size': 34})
    
    if truncate:
        xticks = [i for i in range(0, 901, 100)]
    else:
        xticks = [i for i in range(0, 1600, 200)]
    plt.xticks(xticks, fontsize=34)
    plt.yticks([i for i in range(0, 600, 50)], fontsize=34)
    
    cbar = plt.colorbar(hm)
    cbar.ax.tick_params(labelsize=34)
    
    if savepath:
        plt.savefig(savepath, dpi=300, bbox_inches='tight')
    plt.show()
```

```python
#example for plotting truncated CROCKER plot
dataset = 'C0128'
crocker_path = '../crockers/' + dataset + '_fine_CROCKER.txt'
fed_time = 430
crocker = np.loadtxt(crocker_path, delimiter=',')

plot_experiment_crocker(crocker, fed_time)
```

```python
#example for plotting the full length CROCKER plot
dataset = 'C0128'
crocker_path = '../crockers/' + dataset + '_fine_CROCKER.txt'
fed_time = 430
crocker = np.loadtxt(crocker_path, delimiter=',')

plot_experiment_crocker(crocker, fed_time, truncate=False)
```
